```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the data set credit_card_data
credit_card = pd.read_csv('/content/creditcard.csv')
```

```python
# frist 5 rows data
credit_card.head()
```

```python
credit_card.tail()
```

```python
#dataset informations
credit_card.info()
```

```python
#Checking the number of missing values in each column
credit_card.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transction
credit_card['Class'].value_counts()
```

```python
# separating the data for analysis
legit = credit_card[credit_card.Class == 0]
fraud = credit_card[credit_card.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the value for both transactions
credit_card.groupby('Class').mean()
```

```python
#Number of fraudulent transactions -->492
```

```python
legit_sample = legit.sample(n=492)
```

```python
# Concatenating two data frames
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# splitting the data into features & targets
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
## split the data into training data & testing data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
### model training
```

```python
# logistic model
```

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with training data
model.fit(X_train, Y_train)
```

```python
## Model evaluation
# Accuracy score
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ',training_data_accuracy )
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test data : ', test_data_accuracy)
```
